```python
import pandas as pd
import os.path
import subprocess
import matplotlib
```

```python
if not os.path.exists('analyze_genotypes.tsv'):
    print("grep'ing for genotype counts")
    subprocess.call(["./analyze_genotypes.sh"])
    print("done.")
```

```python
gt_counts=pd.read_csv('./analyze_genotypes.tsv',
                      delimiter='\t',
                      encoding='utf-8',
                      header=None,
                      skiprows=2, # skip header and ALL vcf
                      names=["file","0/0","0/1","1/1","./.","VCF line","Overlapping SVs"]
                      )
print(list(gt_counts.columns.values))
print(gt_counts)
```

```python
bp = gt_counts.groupby("file").plot(kind="bar",x="file",rot=0)
print(bp)
```
